# Write a Python program to get the latest number (total â€“ all countries) of confirmed deaths, recovered and active people with percentage of Novel Coronavirus (COVID-19) cases.

```python
import pandas as pd
covid_data= pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/03-23-2020.csv', usecols = ['Last Update', 'Country/Region', 'Confirmed', 'Deaths', 'Recovered'])
covid_data['Active'] = covid_data['Confirmed'] - covid_data['Deaths'] - covid_data['Recovered']
total_latest_data = covid_data.groupby(['Last Update'])['Confirmed', 'Deaths', 'Recovered', 'Active'].sum()
total_latest_data = total_latest_data.reset_index()
total_latest_data = total_latest_data[total_latest_data['Last Update'] == max(total_latest_data['Last Update'])]
total_latest_data = total_latest_data.reset_index(drop=True)
total_latest_data['Deaths %'] = round(100 * total_latest_data['Deaths'] / total_latest_data['Confirmed'], 2)
total_latest_data['Recovered %'] = round(100 * total_latest_data['Recovered'] / total_latest_data['Confirmed'], 2)
total_latest_data['Active %'] = round(100 * total_latest_data['Active'] / total_latest_data['Confirmed'], 2)
pd.set_option('display.max_columns', None)
print(total_latest_data)
```
